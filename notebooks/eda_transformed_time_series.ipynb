```python
list_of_subgroups = []
with open('../data/final_datasets/time_series/reformated_subgroup_names.txt', 'r') as file:
    for line in file:
        list_of_subgroups.append(line[:-1])
    
for n, sub in enumerate(list_of_subgroups):
    print(n, sub)
    # Load the data
    df = pd.read_csv('../data/final_datasets/time_series/' + sub + '_time_series.csv', 
                     index_col = 'DATE', 
                     parse_dates = True)

    # IQR method to detect outliers
    df = df.select_dtypes(include = ['number']).drop(columns = ['TRANSACTION_ID'])
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    Q3
    IQR = Q3 - Q1
    IQR
    lower_bound = Q1 - 1.5*IQR
    upper_bound = Q3 + 1.5*IQR
    for col in df.columns:
        print(col, df[col].loc[(df[col] < lower_bound[col]) | (df[col] > upper_bound[col])].shape[0]/df.shape[0]*100.0)
    
    print("============================\n")
```
